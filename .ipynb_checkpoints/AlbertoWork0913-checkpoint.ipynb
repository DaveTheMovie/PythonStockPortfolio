```python
#https://www.codingfinance.com/post/2018-03-27-download-price-py/

import pandas as pd
import pandas_datareader as web
import numpy as np
import matplotlib.pyplot as plt
```

```python

aapl = web.get_data_yahoo("AAPL",
                          start = "2017-01-01",
                          end = "2018-03-01")

print(aapl.head())
```

```python

aapl["Adj Close"].plot()
plt.xlabel("Date")
plt.ylabel("Adjusted Price")
plt.title("Apple price data")
plt.show()
```

```python


tickers = ["AAPL", "MSFT", "AMZN", "GS", "JPM"]
prices = web.get_data_yahoo(tickers,
start = "2017-01-01",
end = "2017-01-15")

print(prices.head())
```

```python

prices["Adj Close"].head()
```

```python


prices["Adj Close"].plot()
plt.xlabel("Date")
plt.ylabel("Adjusted Price")
plt.title("All stocks price data")
plt.show()
```
